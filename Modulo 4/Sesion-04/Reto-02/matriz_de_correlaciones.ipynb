```python

mask = np.zeros_like(df.corr())
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
  f, ax = plt.subplots(figsize=(12, 10))
  ax = sns.heatmap(df.corr(),mask=mask,vmin=-1, vmax=1, annot=True, cmap="YlGnBu", linewidths=.5);



  
```
