```python
def escalar_a_01(df):
  df_escalado = (df - df.min()) / (df.max() - df.min())
  return df_escalado
```

```python
def z_score(df):
  df_z_score = (df - df.mean()) / df.std()
  return df_z_score
```

```python
def clip_features(df, iqr_score, clip_top, clip_bottom):
  iqr = df.quantile(0.75) - df.quantile(0.25)
  limite_superior = df.quantile(0.75) + (iqr * iqr_score)
  limite_inferior = df.quantile(0.25) - (iqr * iqr_score)

  new_df = df.copy()

  if clip_top:
    new_df = np.where(df > limite_superior, limite_superior, df)
  if clip_bottom:
    new_df = np.where(df > limite_inferior, limite_inferior, df)
  return new_df
```
